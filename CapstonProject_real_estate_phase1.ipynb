```python
urls=[]
    # scroling down the webpage to get some more images
for _ in range(100):
    driver.execute_script("window.scrollBy(0,1000)")

try:
        all_urls = driver.find_elements(By.XPATH,'//a[@class="HouseList_house_item__yJl0x HouseList_popout_card__11l3u"]')
        for value in all_urls:
            url = value.get_attribute("href")
            if url not in urls:
                urls.append(url)
except NoSuchElementException:
        urls.append("-")
new_len = len(urls)

print(len(urls))    
```

```python
len(urls)
```

```python
# Creating attribute list of Table-1 CSV File
address=[]
city=[]
state_code=[]
country=[]
rental_price=[]
deposit=[]

# Creating  attribute list of Table-2 CSV file

no_of_beds=[]
no_of_bathroom=[]
area=[]
pets_allowed=[]

# Creating  attribute list of Table-3 CSV file

washer_dryer=[]
ac=[]
parking=[]
fireplace=[]
dishwasher=[] 
hardwood_floors=[]
roof_deck=[]
storage=[]


for i in urls:
    driver.get(i)
    time.sleep(3)
    
    # fetching Address
    try:
        address_data1 = driver.find_elements(By.XPATH,'//h1[@class="property-detail_address__G+6zW"]')
        for value in address_data1:
            addr1=value.text
            address_components = addr1.split(', ')
            ad1=address_components[0]
            address.append(ad1)
            ad1_city=address_components[-2]
            city.append(ad1_city)
            ad1_state=address_components[-1]
            state_code.append(ad1_state)
            country.append('India')    # as we can see country=India
    except NoSuchElementException:
        address.append('NaN')
        city.append('NaN')
        state_code.append('NaN')
        country.append('NaN')
        
    
        
    #fetching rental price
    try:
        rental_data = driver.find_elements(By.XPATH,'//h1[@class="property-detail_price__OG8Zu"]')
        for value in rental_data:
            price_inr = value.text
            price_clean = price_inr.replace('â‚¹', '').replace(',', '')
            rental_price.append(price_clean)
            deposit_data=int(price_clean)*2
            deposit.append(deposit_data)
    except NoSuchElementException:
        rental_price.append('NaN')
        deposit.append('NaN')
        
     # fetching no of Beds
    try:
        no_of_beds_data = driver.find_elements(By.XPATH,'/html/body/div[2]/div/div/div/div/div[2]/div[2]/div/div[2]/div/div/div[1]/div[2]/div/div[1]/div/div[2]')
        for value in no_of_beds_data:
            data_bed=value.text
            data_bed1=data_bed.replace('bed', '')
            no_of_beds.append(data_bed1)
            no_of_bathroom.append(data_bed1) #there is no no of bathroom is specify so assuming (no of bathroom = no of bedroom)
    except NoSuchElementException:
        no_of_beds.append('NaN')
    
     
        
    # fetching area 
    try:
        area_data = driver.find_elements(By.XPATH,'/html/body/div[2]/div/div/div/div/div[2]/div[2]/div/div[2]/div/div/div[1]/div[2]/div/div[3]/div/div[2]')
        for value in area_data:
            data_area=value.text
            data_area1=data_area.replace('sqft', '')
            area.append(data_area1)
            pets_allowed.append('No')
    except NoSuchElementException:
        area.append('NaN') 
        
        
    try:
        amenities_ul_list_data = driver.find_elements(By.XPATH,'/html/body/div[2]/div/div/div/div/div[2]/div[2]/div/div[2]/div/div/div[1]/div[4]/div')
        for value in amenities_ul_list_data:
            amenity = value.text
            
            ac_status = 'yes' if "Central Air Conditioning" in amenity else 'no'
            parking_status = 'yes' if "Parking" in amenity else 'no'
            washer_dryer_status = 'yes' if "Washer Dryer In Unit" in amenity else 'no'
            fireplace_status = 'yes' if "Fireplace" in amenity else 'no'
            dishwasher_status = 'yes' if "Dishwasher" in amenity else 'no'
            hardwood_floors_status = 'yes' if "Wood Floors" in amenity else 'no'
            roof_deck_status = 'yes' if "roof deck" in amenity else 'no'
            storage_status = 'yes' if "Storage" in amenity or "Storage Buildings" in amenity else 'no'

            ac.append(ac_status)
            parking.append(parking_status)
            washer_dryer.append(washer_dryer_status)
            fireplace.append(fireplace_status)
            dishwasher.append(dishwasher_status)
            hardwood_floors.append(hardwood_floors_status)
            roof_deck.append(roof_deck_status)
            storage.append(storage_status)
            
            

    except NoSuchElementException:
            
            ac.append('NaN')
            parking.append('NaN')
            washer_dryer.append('NaN')
            fireplace.append('NaN')
            dishwasher.append('NaN')
            hardwood_floors.append('NaN')
            roof_deck.append('NaN')
            storage.append('NaN')
    
```

```python
# Specify the path where your CSV files are located
csv_files = glob.glob('table_1_ahmedabad.csv') + \
             glob.glob('table_1_bengaluru.csv') + \
             glob.glob('table_1_bhopal.csv') + \
             glob.glob('table_1_chandigarh.csv') + \
             glob.glob('table_1_chennai.csv') + \
             glob.glob('table_1_delhi.csv') + \
             glob.glob('table_1_hyderabad.csv') + \
             glob.glob('table_1_kolkata.csv') + \
             glob.glob('table_1_jaipur.csv') + \
             glob.glob('table_1_mumbai.csv') + \
             glob.glob('table_1_deharadun.csv') + \
             glob.glob('table_pune_1.csv') + \
             glob.glob('table_1_patna.csv')

# Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Loop through each CSV file and append its data to the merged_data1 DataFrame
for file in csv_files:
    df = pd.read_csv(file)
    merged_data = merged_data.append(df)

# Add an 'Index' column to the merged_data1 DataFrame
merged_data['Sr_no'] = range(1, len(merged_data) + 1)

# Reorder columns so that 'Sr_no' is the first column
merged_data = merged_data[['Sr_no'] + [col for col in merged_data.columns if col != 'Sr_no']]


# Save the merged DataFrame to a new CSV file
merged_data.to_csv('table_1_data.csv', index=False)

print("CSV files have been successfully merged.")
```
